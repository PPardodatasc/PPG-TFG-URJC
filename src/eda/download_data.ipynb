Due to COVID, datasets since 2020 have not been stored. Only real-time data has been being published. The last dataset form 202 isn't complete. Therefore, data from 2019 is downloaded.

```python
import os 
import requests
from dotenv import load_dotenv
load_dotenv()
```

## Download data

```python
# Data path
data_folder = os.path.join("../..", "data")
os.makedirs(data_folder, exist_ok=True)
data_path = os.path.join(data_folder, "2019_data.zip")
```

```python
# Download 2019 data
data_url = os.environ.get("DATA_2019_URL")

if not data_url:
    print("Error: Variable DATA_2019_URL not found in .env file.")
else:
    try:
        print(f"Starting download of 2019 data in {data_url}. \n It may take a while...\n")
        
        response = requests.get(data_url, stream=True) # stream=True connects to the URL and allows us to download the content in chunks
        response.raise_for_status() 

        with open(data_path, "wb") as file:
            # Iterate over chunks of data and write them to the file
            for i, chunk in enumerate(response.iter_content(chunk_size=8192)): 
                if chunk:
                    print(f"Writing chunk {i} to file...")
                    file.write(chunk)
                    
        print("Download completed and saved to file!")

    except requests.exceptions.HTTPError as eh:
        print(f"HTTP Error: {eh}")
    except requests.exceptions.ConnectionError as ec:
        print(f"Internet/URL conexion error: {ec}")
    except requests.exceptions.Timeout as et:
        print(f"Timeout error: {et}")
    except requests.exceptions.RequestException as e:
        print(f"Unexpected error when calling endpoint: {e}")
```

## Zip to csv

```python
import zipfile

# Open zip and extract files
with zipfile.ZipFile(data_path, "r") as zip_file:
    files = zip_file.namelist()
    print(f"Files inside the ZIP: {files}")
    
    print("Extracting all files...")
    zip_file.extractall(path=data_folder)
    
print("Extraction completed!")
```
