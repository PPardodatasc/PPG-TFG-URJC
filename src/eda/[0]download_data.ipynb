```python
import os 
from dotenv import load_dotenv
load_dotenv()
from utils.utils import download_from_url
```

# Descargar datos

Debido a la COVID-19, los conjuntos de datos desde 2020 no se han almacenado. Solo se han publicado datos en tiempo real. El último conjunto de datos de 2020 está incompleto. Por lo tanto, se han descargado los datos de 2019.

```python
# Dataset paths
data_folder = os.environ.get("DATA_FOLDER_PATH_IPYNB", "../../data/")
os.makedirs(data_folder, exist_ok=True)
data_2019_url = os.environ.get("DATA_2019_URL")
data_2019_path = os.path.join(data_folder, "2019_data.zip") 

spatial_url = os.environ.get("SPATIAL_DATA_URL")
spatial_data_path = os.path.join(data_folder, "spatial_data.parquet") 
```

## Datos históricos de 2019

```python
# Download 2019 data (historical dataset in csv)
if not data_2019_url:
    print("Error: Variable DATA_2019_URL no se ha encontrado en el archivo .env.")
else:
    download_from_url(data_2019_url, data_2019_path)
```

#### Zip a csv

```python
import zipfile

# Open zip and extract files
with zipfile.ZipFile(data_2019_path, "r") as zip_file:
    files = zip_file.namelist()
    print(f"Files inside the ZIP: {files}")
    
    print("Extracting all files...")
    zip_file.extractall(path=data_folder)
    
print(f"Extraction completed to {data_folder}")
```

## Datos espaciales de sensores

```python
# Download spatial data
if not spatial_url:
    print("Error: Variable SPATIAL_DATA_URL not found in .env file.")
else:
    download_from_url(spatial_url, spatial_data_path)
```
